```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

matches_filepath = Path("matches2.csv")
ball_by_ball_filepath  = Path("deliveries1.csv")
ball_by_ball_data = pd.read_csv(ball_by_ball_filepath)
matches_data = pd.read_csv(matches_filepath, index_col="id")

matches_data.loc[matches_data["team1"]=="Rising Pune Supergiant","team1"]="Rising Pune Supergiants"
matches_data.loc[matches_data["team2"]=="Rising Pune Supergiant","team2"]="Rising Pune Supergiants"
matches_data.loc[matches_data["toss_winner"]=="Rising Pune Supergiant","toss_winner"]="Rising Pune Supergiants"
matches_data.loc[matches_data["winner"]=="Rising Pune Supergiant","winner"]="Rising Pune Supergiants"
```

```python
list_stadium = list(matches_data["venue"])
dict_stadium = {}

for stadium in list_stadium:
    if(stadium in dict_stadium):
        dict_stadium[stadium] += 1
    else:
        dict_stadium[stadium] = 1
        
dict_stadium
    
plt.figure(figsize=(15,12))
plt.title("number of matches played on the given venue")
plt.ylabel("name of the stadium",fontweight="bold",fontsize=15)

x=list(dict_stadium.keys())
y=list(dict_stadium.values())

plt.barh(x,y)
```

```python
list_toss_winner = list(matches_data["toss_winner"])
dict_toss_winner = {}

for winner in list_toss_winner:
    if(winner in dict_toss_winner):
        dict_toss_winner[winner] += 1
    else:
        dict_toss_winner[winner] = 1
        
dict_toss_winner

plt.figure(figsize=(15,12))
plt.title("Number of tosses won franchise-wise")
plt.ylabel("franchise name",fontweight="bold",fontsize=15)

x=list(dict_toss_winner.keys())
y=list(dict_toss_winner.values())

plt.barh(x,y)
```

```python
list_toss_decision = list(matches_data["toss_decision"])
list_toss_decision
dict_toss_decision = {}

for decision in list_toss_decision:
    if(decision in dict_toss_decision):
        dict_toss_decision[decision] += 1
    else:
        dict_toss_decision[decision] = 1
        
dict_toss_decision

plt.figure(figsize=(12,8))
plt.title("toss dicision",fontweight='bold',fontsize=20)
plt.pie(dict_toss_decision.values(),labels=dict_toss_decision.keys())
```

```python
list_moms = list(matches_data["player_of_match"])
 
dict_moms = {}

for player in list_moms:
    if(player in dict_moms):
        dict_moms[player] += 1
    else:
        dict_moms[player] = 1
        
sorted_moms = {}     

sorted_moms_keys = sorted(dict_moms, key = dict_moms.get)

for w in sorted_moms_keys:
    sorted_moms[w] = dict_moms[w]


sorted_moms



 
 
```

```python
top_moms_names = list(sorted_moms.keys())[-10:-1]
top_moms_no = list(sorted_moms.values())[-10:-1]
last = len(list(sorted_moms.keys()))
top_moms_names.append(list(sorted_moms.keys())[last-1])
top_moms_no.append(list(sorted_moms.values())[last-1])
 
plt.figure(figsize=(12,8))
plt.title("maximum man of the match awards",fontsize=20,fontweight='bold')
plt.xlabel("number of awards",fontweight='bold',fontsize=15)
plt.ylabel("name of the player",fontweight='bold',fontsize=15)
plt.barh(top_moms_names,top_moms_no)
```

```python

name = 'Yuvraj Singh'
df = pd.Series({'Innings':0,'Runs':0,'Highest Score':0,'Average':0,'Strike Rate':0})
X = ball_by_ball_data.loc[(ball_by_ball_data.batsman==name)]
innings = len(X['match_id'].unique())
runs = ball_by_ball_data.groupby('batsman').batsman_runs.sum().loc[name]
    
X1=ball_by_ball_data.loc[(ball_by_ball_data.batsman==name)]
highest_Score=X1.groupby('match_id').batsman_runs.sum().sort_values(ascending=False).iloc[0]
    
outs = 0
for i in range(len(ball_by_ball_data)):
    if(ball_by_ball_data.loc[i,'player_dismissed']==name):
        outs+=1
average = runs/float(outs)
balls = len(X1) - len(X1.loc[X1.wide_runs != 0])
strike_rate = (runs/ float(balls))*float(100)

 
df.loc[ 'Innings'] = innings
df.loc ['Runs'] = runs
df.loc['Highest Score'] = highest_Score
df.loc['Average'] = average
df.loc['Strike Rate'] = strike_rate
df
```

```python
name = 'YS Chahal'
df = pd.Series({'Matches':0,'Wickets':0,'Average': 0,'Best figure':(0,0),'Economy':0,"Strike Rate":0})
Y = ball_by_ball_data.loc[(ball_by_ball_data.bowler==name)]
innings = len(Y['match_id'].unique())
Y1 = ball_by_ball_data.groupby('bowler').dismissal_kind.count().loc[name]
Y2 = ball_by_ball_data.loc[ball_by_ball_data.bowler == name ]
Y3 = Y2.loc[Y2.dismissal_kind=='run out']
wickets = Y1 - len(Y3)
runs = ball_by_ball_data.groupby('bowler').total_runs.sum().loc[name]
best_wickets= ball_by_ball_data.groupby(['bowler','match_id']).count().sort_values(by='player_dismissed',ascending= False).loc[name].iloc[0,16]
match_id_best = list(ball_by_ball_data.groupby(['bowler','match_id']).count().sort_values(by='player_dismissed',ascending= False).loc[name].index)[0] 
best_runs = ball_by_ball_data.groupby(['bowler','match_id']).total_runs.sum().loc[name].loc[match_id_best]
overs_bowled = (ball_by_ball_data.loc[ball_by_ball_data.bowler == name].groupby('match_id').over.count().sum())/6
economy = runs/overs_bowled
sr = (overs_bowled*6)/wickets
average = runs/wickets

df.loc['Matches'] = innings
df.loc['Wickets'] = wickets
df.loc['Average'] = average
df.loc['Best figure'] = (best_wickets,best_runs)
df.loc['Economy'] = economy
df.loc['Strike Rate'] = sr

df
```
